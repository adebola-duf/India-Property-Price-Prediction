```python
def missing_dataframe(df):
    data_columns = df.columns
    missing_total = df.isnull().sum()
    missing_percentage = 100 * missing_total / df.shape[0]

    missing_df = pd.concat([missing_total, missing_percentage], axis=1)
    missing_df.rename(columns={0: "# Missing", 1: "% Missing"}, inplace=True)
    missing_df = missing_df.sort_values(by=["% Missing"], ascending=False)
    missing_df = missing_df.loc[missing_df["% Missing"] > 0, :]

    # Print some summary information
    print("Your selected dataframe has " + str(df.shape[1]) + " columns.\n" +
          "There are " + str(missing_df.shape[0]) +
          " columns that have missing values.")

    return missing_df


missing_df = missing_dataframe(df)
missing_df
```

```python
# so some of the columns have excessively too many missing values so i'mma just start by dropping those ones first i.e > 90
df.drop(
    columns=missing_df.loc[missing_df["% Missing"] > 90, :].index, inplace=True)
df
```

```python
from scipy import stats

plt.style.use('fivethirtyeight')


def three_charts(series):
    fig = plt.figure(figsize=(12, 8))
    grid = plt.GridSpec(nrows=2, ncols=3, wspace=0.4, hspace=0.3)

    ax1 = fig.add_subplot(grid[0, :2])
    ax2 = fig.add_subplot(grid[:, 2])
    ax3 = fig.add_subplot(grid[1, :2])

    ax1.set_title("Histogram X KDE")
    ax1.set_xlabel(series.name)
    sns.distplot(x=series, ax=ax1)

    ax2.set_title("Boxplot")
    sns.boxplot(y=series, ax=ax2)

    ax3.set_title('QQ Plot')
    stats.probplot(series, plot=ax3)
```
